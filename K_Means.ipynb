```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Load the iris dataset
iris = load_iris()
X = iris.data[:, 2:4]  # Consider petal length and width

# Create a KMeans model with 3 clusters
kmeans_model = KMeans(n_clusters=3, random_state=0)

# Fit the model and predict clusters
clusters = kmeans_model.fit_predict(X)

# Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis')
plt.scatter(kmeans_model.cluster_centers_[:, 0], kmeans_model.cluster_centers_[:, 1], c='red', marker='x', s=100)
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.title('K-means Clustering of Iris Dataset (Petal Length and Width)')
plt.show()
```
